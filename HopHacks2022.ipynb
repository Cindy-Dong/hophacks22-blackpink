# Function to Calculuate Attendence Percentage

```python
def calculate_attendence(df, name):
    for i in range(len(df)):
        if (df.loc[i].at['Name'] == name):
            percentage = df.loc[i].at['Students_Present'] / df.loc[i].at['Students_Enrolled']
            df.loc[i, 'Attendence_Percentage'] = percentage
```

```python
calculate_attendence(PILOT_df, 'Lisa')
calculate_attendence(PILOT_df, 'Jennie')
calculate_attendence(PILOT_df, 'Jisoo')
calculate_attendence(PILOT_df, 'Rose')
PILOT_df
```

# Function to Calculuate Salary

```python
def calculate_salary(df, name):
    base = 15
    for i in range(len(df)):
        if (df.loc[i].at['Name'] == name):
            salary = base * df.loc[i].at['Hours'] * (1 + df.loc[i].at['Experience_Level'] * 0.05)
            df.loc[i, 'Salary'] = salary
```

```python
calculate_salary(PILOT_df, 'Lisa')
calculate_salary(PILOT_df, 'Jennie')
calculate_salary(PILOT_df, 'Jisoo')
calculate_salary(PILOT_df, 'Rose')
PILOT_df
```

# Calculate Sum of Salary

```python
def sum_salary(df, name):
    salary = 0
    for i in range(len(df)):
        if (df.loc[i].at['Name'] == name):
            salary += df.loc[i].at['Salary']
    return salary
```

```python
name = input("Enter your name to get your total salary: ")
s = sum_salary(PILOT_df, name)
print("Your current total salary is: " + str(s))
```

# Filter a Certain Person

```python
person_df = pd.DataFrame()
person_df
```

```python
def filter_person(df, person_df, name):
    person_df = pd.DataFrame()
    person_df = df[df.Name == name]
    person_df = person_df.sort_values(by=['Date'])
    person_df = person_df.reset_index()
    person_df = person_df.drop(['index'], axis=1)
    return person_df
```

```python
name = input("Enter your name to view your personal work progress track: ")
person_df = filter_person(PILOT_df, person_df, name)
person_df
```

# Plot Attendence Percentage

```python
def plot_attendence(person_df, name):
    lines = person_df.plot.line(x='Date', y='Attendence_Percentage')
    lines.set_title('Attendence Percentage Track for ' + name + ' ')
```

```python
name = input("Enter your name to get your attendence percentage track: ")
plot_attendence(person_df, name)
```
