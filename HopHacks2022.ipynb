```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [['Lisa', 4, 'Calculus III', 20, 15, 10, 0], ['Jennie', 3, 'Introductory Chemistry', 25, 20, 15, 0], ['Jisoo', 2, 'General Physics', 30, 15, 10, 0], ['Rose', 1, 'Linear Algebra',35, 20, 15, 0]]
PILOT_df = pd.DataFrame(data, columns=['Name','Experience_Level','Course','Hours','Students_Enrolled','Students_Present','Salary'])
PILOT_df
```

```python
Attendence_Percentage = [0]*4
PILOT_df['Attendence_Percentage'] = Attendence_Percentage
PILOT_df
```

```python
def calculate_attendence(df, name):
    for i in range(len(df)):
        if (df.loc[i].at['Name'] == name):
            percentage = df.loc[i].at['Students_Present'] / df.loc[i].at['Students_Enrolled']
            df.loc[i, 'Attendence_Percentage'] = percentage
```

```python
calculate_attendence(PILOT_df, 'Lisa')
calculate_attendence(PILOT_df, 'Jennie')
calculate_attendence(PILOT_df, 'Jisoo')
calculate_attendence(PILOT_df, 'Rose')
PILOT_df
```

```python
def calculate_salary(df, name):
    base = 15
    for i in range(len(df)):
        if (df.loc[i].at['Name'] == name):
            salary = base * df.loc[i].at['Hours'] * (1 + df.loc[i].at['Experience_Level'] * 0.05)
            df.loc[i, 'Salary'] = salary
```

```python
calculate_salary(PILOT_df, 'Lisa')
calculate_salary(PILOT_df, 'Jennie')
calculate_salary(PILOT_df, 'Jisoo')
calculate_salary(PILOT_df, 'Rose')
PILOT_df
```

```python
def collect_info(df):
    name = input("What is your name? ")
    experience_level = input("What is your current experience level? ")
    course = input("What course do you lead? ")
    hours = input("How many hours did you work today? ")
    students_enrolled = input("How many students enrolled in your session? ")
    students_present = input("How many students attended today? ")
    
    new_df = pd.DataFrame([[name, experience_level, course, hours, students_enrolled, students_present, 0, 0]])
    df.append(new_df, ignore_index=True)
```

```python
collect_info(PILOT_df)
PILOT_df
```
